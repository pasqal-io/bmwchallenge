```python
import torch
import torch.nn as nn
import torch.nn.functional as F


original_model = models.resnet18(pretrained=True)

class CustomResNet(nn.Module):
    def __init__(self):
        super(CustomResNet, self).__init__()
        self.features = nn.Sequential(
            *list(original_model.children())[:-1]
        )
        self.final_layer = nn.Linear(in_features=512, out_features=1, bias=True)
        
    def forward(self, x):
        x = self.features(x)
        x = x.squeeze(2)
        x = x.squeeze(2)
        x = self.final_layer(x).squeeze(1)
        return x
```

```python
custom_resnet = CustomResNet()
epochs = 10

optimizer = torch.optim.Adam(custom_resnet.parameters(), lr=0.01)

for images, labels in dataloader:
    t0 = time()
    output = custom_resnet(images)
    loss = F.binary_cross_entropy_with_logits(output, labels.float())
    loss.backward()
    optimizer.step()
    t1 = time()
    print(t1-t0)
```
